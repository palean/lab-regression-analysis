## Challenge 1
I work at a coding bootcamp, and I have developed a theory that the younger my students are, the more often they are late to class. In order to test my hypothesis, I have collected some data in the following table:

| StudentID | Age | Tardies |
|--------|-----|------------|
| 1      | 17  | 10         |
| 2      | 51  | 1          |
| 3      | 27  | 5          |
| 4      | 21  | 9         |
| 5      | 36  |  4         |
| 6      | 48  |  2         |
| 7      | 19  |  9         |
| 8      | 26  | 6          |
| 9      | 54  |  0         |
| 10     | 30  |  3         |

Use this command to create a dataframe with the data provided in the table. 
~~~~
student_data = pd.DataFrame({'X': [x_values], 'Y': [y_values]})
~~~~

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm #this is for linear regressions

age = [17,51,27,21,36,48,19,26,54,30]
tardies = [10,1,5,9,4,2,9,6,0,3]
student_data = pd.DataFrame()
student_data['age'] = age
student_data['tardies'] = tardies
student_data
```

Draw a dispersion diagram (scatter plot) for the data.

```python
# con Seaborn

plt.rcParams['figure.figsize'] = (22,12)
sns.scatterplot('age','tardies',data=student_data, size='age',sizes=(300,600), color='green')
plt.xlabel('Age', size=30)
plt.ylabel('Tardies', size =30)
plt.title('Dependence of tardiness according to age', size =30)
plt.show()

#Líneas con mathlib
#params = {'figure.figsize':(22,12),
 #         'legend.fontsize':20,
 #         'axes.labelsize':'xx-large', #x-large is a global size
 #         'axes.titlesize':'xx-large',
 #         'xtick.labelsize':'xx-large',
 #         'ytick.labelsize':'xx-large'}
#plt.rcParams.update(params)
#plt.title('Dependence of tardiness according to age',fontsize=30)
#plt.scatter(student_data['X'],student_data['Y'])
#plt.show();
```

Do you see a trend? Can you make any hypotheses about the relationship between age and number of tardies?

Calculate the covariance and correlation of the variables in your plot. What is the difference between these two measures? Compare their values. What do they tell you in this case? Add your responses as comments after your code.

```python
print(f'Correlation between age and tardies: {student_data.corr()}')
#There is a linear relationship between the age and the tardies variables, becouse their proximity to 1

print(f'Covariance between age and tardies: {student_data.cov()}')
#What does it means?
```

Build a regression model for this data. What will be your outcome variable? What type of regression are you using? Add your responses as comments after your code.

```python
X = sm.add_constant(student_data['age'])
Y = student_data['tardies']

regresion = sm.OLS(Y,X).fit()
prediccion = regresion.predict(X)

print(regresion.summary());
```

Plot your regression model on your scatter plot.

```python
sns.lmplot('age','tardies',data=student_data, fit_reg=True,order=1)
```

Interpret the results of your model. What can conclusions can you draw from your model and how confident in these conclusions are you? Can we say that age is a good predictor of tardiness? Add your responses as comments after your code.

## Challenge 2
For the second part of this lab, we will use the vehicles.csv data set. You can find a copy of the dataset in the git hub folder. This dataset includes variables related to vehicle characteristics, including the model, make, and energy efficiency standards, as well as each car's CO2 emissions. As discussed in class the goal of this exercise is to predict vehicles' CO2 emissions based on several independent variables. 

```python
vehicles_df = pd.read_csv('vehicles.csv')
vehicles_df.head()
```

Let's use the following variables for our analysis: Year, Cylinders, Fuel Barrels/Year, Combined MPG, and Fuel Cost/Year. We will use 'CO2 Emission Grams/Mile' as our outcome variable. 

Calculate the correlations between each of these variables and the outcome. Which variable do you think will be the most important in determining CO2 emissions? Which provides the least amount of helpful information for determining CO2 emissions? Add your responses as comments after your code.

```python
# Correlation analysis

vehicles_corr = vehicles_df[['Year', 'Cylinders', 'Fuel Barrels/Year', 'Combined MPG', 'Fuel Cost/Year', 'CO2 Emission Grams/Mile']].corr().sort_values(by=['CO2 Emission Grams/Mile'],ascending=False)['CO2 Emission Grams/Mile']
vehicles_corr
```

Build a regression model for this data. What type of regression are you using? Add your responses as comments after your code.

```python
X = sm.add_constant(vehicles_df[['Cylinders', 'Fuel Barrels/Year', 'Combined MPG', 'Fuel Cost/Year']]) #taking out the Year

Y = vehicles_df['CO2 Emission Grams/Mile']

regresion = sm.OLS(Y,X).fit()
prediccion = regresion.predict(X)

print(regresion.summary());
```

Print your regression summary, and interpret the results. What are the most important varibles in your model and why? What can conclusions can you draw from your model and how confident in these conclusions are you? Add your responses as comments after your code.

```python
# To know the most important variables, here are graphics showing relation between the variables and the Y ('CO2 Emission Grams/Mile')

def yGraphic(Cylinders, FuelBarrelsYear, CombinedMPG, FuelCostYear):
    return 131.1563+(1.0448*Cylinders)+(19.6271*FuelBarrelsYear)+(-3.1666*CombinedMPG)+(0.0294*FuelCostYear)

sns.lmplot('Cylinders', 'CO2 Emission Grams/Mile', data=vehicles_df, fit_reg=True, order=1)
plt.xlabel('Cylinders',size=15)
plt.ylabel('CO2 Emission Grams/Mile',size=15)
plt.title('How the number of cylinders explains the CO2 Emission Grams/Mile',size=15)
plt.show()

#######

x2 = np.arange(0,max(vehicles_df['Fuel Barrels/Year'])+1,1)
plt.plot(x2,yGraphic(0,x2,0,0))                    #regresion graph

colors = np.random.rand(vehicles_df.shape[0])
plt.scatter(vehicles_df['Fuel Barrels/Year'],vehicles_df['CO2 Emission Grams/Mile'], c=colors, alpha=0.5)
plt.xlabel('Fuel Barrels/Year',size=15)
plt.ylabel('CO2 Emission Grams/Mile',size=15)
plt.title('How the Fuel Barrels/Year explains the CO2 Emission Grams/Mile',size=15)
plt.axis([vehicles_df['Fuel Barrels/Year'].min(),vehicles_df['Fuel Barrels/Year'].max(),vehicles_df['CO2 Emission Grams/Mile'].min(),vehicles_df['CO2 Emission Grams/Mile'].max()])

plt.show()

#La variable de Fuel barrels per year tiene una mejor regresión que minimiza los coeficientes de error, comparado con el número de cilindros.
# Sin embargo, ésta regresión se podría mejorar.
```

We want to know if the relationship modeled by the two random variables is linear or not, and therefore if it is appropriate to model it with a linear regression. 
First, build a dataframe with the data. 

```python
party = pd.DataFrame({'invitados': [1,3,4,6,8,9,11,14], 
                      'asistentes': [1,2,4,4,5,7,8,13]})
party
```

Draw a dispersion diagram (scatter plot) for the data, and fit a regression line.

```python
from sklearn.linear_model import LinearRegression

X = party['invitados'].values.reshape(-1,1) #ajustando dimensiones
Y = party['asistentes'].values.reshape(-1,1) #ajustando dimensiones

regresion = LinearRegression()              #regresión
regresion.fit(X,Y)                          #regresión
prediccion = regresion.predict(X)

plt.scatter(X,Y)
plt.plot(X,prediccion,color='red')
plt.show()
```
